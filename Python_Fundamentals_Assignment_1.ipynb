```python
import sympy as sym # to create and manipulate symbolic expressions
import random # to generate a random number x0

def devesh_sharma_NR(a, b, c, tolerance, max_iterations):
    """
    A function that finds the root of an equation using the Newton-Raphson method
    Parameters:
        coefficients: a list of numbers representing the coefficients of the polynomial function
        tolerance: a positive number representing the desired accuracy of the root
        max_iterations: a positive integer representing the maximum number of iterations allowed
    Returns:
        root: a number representing the approximate root of the equation
        error: a number representing the absolute error of the root
        iterations: an integer representing the number of iterations performed
    Raises:
        ValueError: if coefficients is empty or not a list, or if tolerance or max_iterations is not positive
        ZeroDivisionError: if the derivative of the polynomial is zero at any point
    """
    # Validate the input parameters
    if not isinstance(coefficients, list) or len(coefficients) == 0:
        raise ValueError("coefficients must be a non-empty list")
    if not isinstance(tolerance, (int, float)) or tolerance <= 0:
        raise ValueError("tolerance must be a positive number")
    if not isinstance(max_iterations, int) or max_iterations <= 0:
        raise ValueError("max_iterations must be a positive integer")

    # Define the symbols for x and err
    x = sym.Symbol('x')
    err = sym.Symbol('err')

    # Define the polynomial function using sympy
    f = a*x**2 + b*x + c
    f_prime = f.diff(x) #To find the derivative
    x0 = random.randint(-100, 100) # For simplicity taking random number between -100 and 100
    counter = 0
    error = tolerance + 1 # to ensure that error is greater than tolerance initially
    root = None
    while error > tolerance and counter < max_iterations: # Loop until error is less than tolerance or counter exceeds max_iterations
        f_x0 = sym.N(f.subs(x, x0))
        f_prime_x0 = sym.N(f_prime.subs(x, x0))
        if f_prime_x0 == 0: # Check if f'(x0) is zero and raise an exception if so
            raise ZeroDivisionError("The derivative of the polynomial is zero at x = {}".format(x0))
        x1 = x0 - f_x0 / f_prime_x0 # Compute x1 using the Newton-Raphson formula: x1 = x0 - f(x0) / f'(x0)
        error = abs(x1 - x0)
        x0 = x1
        counter += 1
    return x0, error, iterations


if __name__ == "__main__":
    a= int(input("Enter Coefficient-1"))
    b= int(input("Enter Coefficient-2"))
    c=int(input("Enter Coefficient-3"))
    tolerance = 1e-6 # desired accuracy of the root
    max_iterations = 100 # Fixing maximum number of iterations

    try:
        root, error, iterations = devesh_sharma_NR(a, b, c, tolerance, max_iterations)
        print("The root of the equation is approximately {:.6f}".format(root))
        print("The absolute error of the root is {:.6f}".format(error))
        print("The number of iterations performed is {}".format(iterations))
    except Exception as e: # Print any exception that occurs during execution
        print(e)
```

```python

def devesh_sharma_NRsqrt(number, tolerance, max_iterations):
    if not isinstance(number, (int, float)) or number <= 0:
        raise ValueError("number must be a positive number")
    if not isinstance(tolerance, (int, float)) or tolerance <= 0:
        raise ValueError("tolerance must be a positive number")
    if not isinstance(max_iterations, int) or max_iterations <= 0:
        raise ValueError("max_iterations must be a positive integer")

    x0 = number / 2 # Just for fun x0 as half of the number
    error = float("inf") # for setting initial value of error
    iterations = 0
    while error > tolerance and iterations < max_iterations:
        fx0 = x0**2 - number # Evaluate f(x0) and f'(x0) using the formula for square root: f(x) = x**2 - number and f'(x) = 2*x
        f_prime_x0 = 2*x0
        x1 = x0 - fx0 / f_prime_x0
        error = abs(x1 - x0)
        x0 = x1
        iterations += 1
    return x0, error, iterations #x0 = root; error = absolute error; iterations = number of iter

if __name__ == "__main__": # Test the function with an example number: 25
  number = int(input("Enter number"))
  tolerance = 1e-6
  max_iterations = 100

  try:
      root, error, iterations = devesh_sharma_NRsqrt(number, tolerance, max_iterations)
      print("The square root of {} is approximately {:.6f}".format(number, root))
      print("The absolute error of the square root is {:.6f}".format(error))
      print("The number of iterations performed is {}".format(iterations))
  except Exception as e: #Exception handling
      print(e)
```
